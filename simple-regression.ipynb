### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Dataset

```python
data = pd.read_csv('Salary_dataset.csv')
```

```python
data.head()
```

```python
data = data.drop(columns=['Unnamed: 0'])
```

### Scatter Plot

```python
plt.scatter(data['YearsExperience'], data['Salary'])
plt.title('Scatter Plot of Salary and Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary');
```

```python
def gradient_descent(current_m, current_b, points, learning_rate):
    m_gradient = 0
    b_gradient = 0
    
    n = len(points)
    
    for i in range(n):
        x = points.iloc[i].YearsExperience
        y = points.iloc[i].Salary
        
        prediction = current_m * x + current_b
        m_gradient += -(2/n) * x * (y - prediction)
        b_gradient += -(2/n) * (y - prediction)
        
    m = current_m - m_gradient * learning_rate
    b = current_b - b_gradient * learning_rate
    
    return m, b        
```

```python
m = 0
b = 0
learning_rate = 0.001
epochs = 1000
```

```python
for i in range(epochs):
    if i % 100 == 0:
        print(f"Epoch: {i}")
    m, b = gradient_descent(m, b, data, learning_rate)

print(m, b)
```

```python
plt.scatter(data['YearsExperience'], data['Salary'])
plt.plot(list(range(len(data))), [m * x + b for  x in range(len(data))], color="red")
plt.title('Regression line for Salary and Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary');
```
